```python
import openai
import json
import time

openai.api_key = 'myapikey' #openai API key

file = "test_HW3dataset.json"

with open(file, newline='',encoding="utf-8") as jsonfile:
    data = json.load(jsonfile)

ans = []

for i in data:
    text = i[0]
    for j in i[1]:
        prompt = str(text)+'\n'+str(j)+'\n give me the index of the correct answer.' 
        completions = openai.Completion.create(
            engine="text-davinci-002",
            prompt=prompt,
            max_tokens=1024,
            n=1,
            stop=None,
            temperature=0.5,
        )
        response = completions.choices[0].text
        print('prompt:')
        print(prompt)
        print('response:')
        print(response)
        ans.append(response)
        time.sleep(3)
```

```python
rans = []
aans = 0
for i in ans:
    for j in i:
        if j in ['0','1','2','3']:
            aans = int(j)+1
            break
    rans.append(aans)
```

```python
import csv
    
with open('top1021.csv','a',newline = '') as fd:
    writer = csv.writer(fd)
    writer.writerow(['index','answer'])
    
    
for index,pred in enumerate(rans):
    with open('top1021.csv','a',newline = '') as fd:
        writer = csv.writer(fd)
        writer.writerow([index,pred])
```
